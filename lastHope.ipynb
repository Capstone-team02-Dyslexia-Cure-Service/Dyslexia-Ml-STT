```python
import json
import os
import re
import shutil
import zipfile
```

```python
def rule(x):
    # 괄호
    a = re.compile(r'\([^)]*\)')
    # 문장 부호
    b = re.compile('[^가-힣0-9 ]')
    x = re.sub(pattern=a, repl='', string= x)
    x = re.sub(pattern=b, repl='', string= x)
    return x
```

```python
def count_unique_korean_words(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()

    # 한글 단어만 추출 (정규 표현식 이용)
    words = re.findall(r'\b[가-힣]+\b', text)

    # 중복을 제거하기 위해 set 사용
    unique_words = set(words)

    return len(unique_words)

# 파일 경로를 지정해주세요
file_path = '/content/textfile/train_zero.txt'

unique_word_count = count_unique_korean_words(file_path)
print(f'유일한 한글 단어의 개수: {unique_word_count}')
```

```python
import chardet

def detect_file_encoding(file_path):
    with open(file_path, 'rb') as file:
        raw_data = file.read()
        result = chardet.detect(raw_data)
    encoding = result['encoding']
    confidence = result['confidence']
    return encoding, confidence

# 파일 경로를 지정해주세요
file_path = '/content/textfile/train_zero.txt'

encoding, confidence = detect_file_encoding(file_path)
print(f"파일 인코딩: {encoding} (신뢰도: {confidence*100:.2f}%)")
```

```python
def count_lines(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        line_count = sum(1 for line in file)
    return line_count

# 파일 경로를 지정해주세요
file_path = 'path_to_your_file.txt'

total_lines = count_lines(file_path)
print(f'총 줄 수: {total_lines}')
```
